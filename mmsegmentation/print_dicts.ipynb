```python
from mmseg.apis import init_model
```

```python
config_file = '/home/robert.breslin/alessandro/testing/paper_2/mmsegmentation/configs/sebnet/sebnet_ablation40_1xb6-160k_cityscapes.py'
model_2 = init_model(config_file)
```

```python
print(model_2.state_dict().keys())
```

```python
# Define the mapping from old prefixes to new prefixes
key_mapping = {
    'decode_head.fusion': 'decode_head.pre_head',
    'decode_head.sbd.': 'decode_head.edge_module.',
    'decode_head.d_head': 'decode_head.hed_head',
    'decode_head.d_cls_seg': 'decode_head.hed_cls_seg'
}

# Load the original state dictionary
# Replace 'path_to_your_checkpoint.pth' with the actual file path
state_dict = model_1['state_dict'].copy()

# Create a new state dictionary with remapped keys
new_state_dict = {}
for old_key, value in state_dict.items():
    new_key = old_key
    for old_prefix, new_prefix in key_mapping.items():
        if old_key.startswith(old_prefix):
            # Replace the old prefix with the new one
            new_key = new_prefix + old_key[len(old_prefix):]
            break  # Assuming no overlapping prefixes, stop after first match
    new_state_dict[new_key] = value

# Now you can use new_state_dict, e.g., load it into your model
# model.load_state_dict(new_state_dict)

# Optionally, save the remapped state dict
# torch.save(new_state_dict, 'remapped_checkpoint.pth')
```

```python
model_3 = model_1.copy()
model_3['state_dict'] = new_state_dict
torch.save(model_3, '/home/robert.breslin/alessandro/testing/paper_2/mmsegmentation/work_dirs/sebnet_baseline-p-d-sbd-bas-head_2xb6_mapillaryv2/20250916_135803/checkpoints/remapped_checkpoint.pth')
```
